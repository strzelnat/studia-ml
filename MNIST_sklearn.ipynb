```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import time
```

```python
import pandas as pd
```

```python
results = pd.DataFrame(columns = ('klasyfikator', 'ocena', 'czas_uczenia', 'czas_testu'))
```

```python
def run(xtr, ytr, xts, yts, clf, clf_name):
  global results
  s = time.time()
  x=clf.fit(xtr,ytr)
  time_train = time.time() - s
  s = time.time()
  score_tst = x.score(xts, yts)
  time_test = time.time() - s
  print('ocena=%0.4f,czas uczenia=%8.3f,czas_testu=%8.3f' %(score_tst,time_train,time_test))
  new_row = pd.DataFrame({'klasyfikator':[clf_name], 'ocena':[score_tst], 'czas_uczenia':[time_train], 'czas_testu':[time_test]})

  results = pd.concat([results, new_row], ignore_index=True)
```

```python
from sklearn.neighbors import NearestCentroid, KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
```

```python
clfs = [NearestCentroid(), KNeighborsClassifier(), GaussianNB(), MultinomialNB(), LogisticRegression(), DecisionTreeClassifier(), RandomForestClassifier(), MLPClassifier()]
```

```python
start = time.time()
for i in clfs:
  print(i)
  run(trainX, trainY, testX, TestY, i, str(i))
  run(trainX_01, trainY, testX_01, TestY, i, str(i) + '_01')
  if str(i)[:13]!='MultinomialNB':
    run(trainX_std, trainY, testX_std, TestY, i, str(i) + '_std')
end = time.time()
print(f'Czas:{end - start}')
```
